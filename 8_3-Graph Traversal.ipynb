```python
graph = {'a': ['b', 'd'],'b': ['a', 'd'], 'c': ['d'], 'd': ['a', 'b', 'c']}
def findPath(graph,start,end,path=[]):
    path = path + [start]
    for node in graph[start] :
        if not node in path :
            newPath = findPath(graph, node, end, path)
            if newPath:
                return newPath
    if start == end :
        return path
    if not start in graph:
        return None
    return None

print(findPath(graph,'a','b'))
```

```python
import networkx as nx
graph1 = {'a':['b','c','d'],'b':['a','e','c'],'c':['d','a','b','e'],'d':['a','e','c'],'e':['d','c','b']}
G1 = nx.Graph()
for i in graph1.keys():
    G1.add_node(i)
    for j in graph1[i]:
        G1.add_edge(i,j)
jalur = findPath(graph1,'a','e')
print(jalur)
color = ['red' if i[0] in jalur and i[1] in jalur else 'black' for i in G1.edges()]
colorNode = ['red' if i in jalur else 'lightblue'  for i in G1.nodes()]
nx.draw(G1, with_labels = True, edge_color=color,node_color=colorNode)

```
