- Data yang memiliki nilai besar, diletakkan terlebih dahulu di indek-indeks akhir, sehingga pada akhir iterasi indeks-indeks akhir ditempati data yang memiliki nilai besar.<br><br>
- Contoh Proses perbandingan yang dilakukan pada modifikasi algoritma selection sort ini dan outputyang harus dihasilkan pada fungsi yang akan dibuat, dapat dilihat pada Gambar 3


Gambar 3: Selection Sort - Ascending

```python
# Jawaban 

def selectionSortAscending(data):
    print('Data = ',data)
    for i in range(len(data)-1,-1,-1):
        idxMax = i
        for j in range(i-1,-1,-1):
            if data[idxMax] < data[j]:
                idxMax=j
        data[i],data[idxMax]=data[idxMax],data[i]
        print(f"data[{idxMax}] is the max value, swap data[{idxMax}] and data[{i}],  New Data : {data}")
    print('Sorted Data =',data)  
    
a = [10,7,6,5,3,1,0,-5]
selectionSortAscending(a)
```

Kerjakan untuk semua mahasiswa.<br>
Modifikasi algoritma Bubble Sort agar iterasi terhenti ketika tidak ada lagi data yang akan ditukar posisinya
(semua data sudah berada di tempat yang tepat). Contoh Bubble Sort tanpa modifikasi dapat dilihat pada
5a, sedangkan modifikasi bubble sort dapat dilihat pada 5b.

(a)	Bubble Sort Modifikasi

```python
# Jawaban

def modifiedBubbleSort(data):
    i = len(data)-1
    while i > 0 :
        print('Iterasi ke-',len(data)-i,'Jumlah Iterasi-',i)
        temp=0
        for j in range(i):
            if data[j+1] < data [j] :
                data[j],data[j+1]=data[j+1],data[j]
                temp+=1
            print(j+1,'=',data)
        if temp == 0 :
            i = 0
        i-=1

b = [10,2,5,8,1,20,2,2,4]
modifiedBubbleSort(b)
```
