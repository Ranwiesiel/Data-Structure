Buatlah Fungsi dan code yang diperlukan untuk memanggil fungsi tersebut dengan menggunakan bahasa Python, antara lain :
- Fungsi createList untuk membentuk suatu list dengan parameter atau argumen berupa ukuran list 
- Fungsi untuk menghitung Persamaan (2) yaitu :
$$ y=\sum_{i=0}^{n-1} x_i $$ 
$$ P_i=\frac{1}{y}x_i $$
dimana n adalah jumlah data yang terdapat pada list (dari fungsi createList), xi adalah data ke-i dari suatu list, dan Pi adalah probabilitas dari data xi. 
<br><br>
- Fungsi untuk menghitung Persamaan Gini, seperti yang dituliskan pada Persamaan (3) yaitu:
$$ gini=1-\sum_{i=0}^{n-1}(P_i)^2 $$
dimana n adalah jumlah data yang terdapat pada list (dari fungsi createList), dan xi adalah data ke-i dari suatu list, dan Pi adalah probabiltas pada Persamaan(2) contoh output dapat dilihat pada Gambar 3



```python
def createList(size):
    data = []
    for i in range(size):
        inp = int(input("Masukkan data : "))
        data.append(inp)
    return data
def sigmaY(data):
    jml = 0
    for i in data :
        jml += i
    return jml
def probabilitas(data,y):
    result = [(1/y)*i for i in data]
    print(result)
    return result
def gini(pi):
    hasil = 0
    for j in pi :
        hasil += j**2
    return 1 - hasil
def gini(pi):
    result = 0
    for i in pi :
        result+=i**2
    return 1 - result
```

```python
x = createList(6)
y = sigmaY(x)
pi = probabilitas(x,y)
gini = gini(pi)
print("x = ",x," ; y =",y)
print("probabilitas = ",pi)
print("Gini = ",gini)
```
