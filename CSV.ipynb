```python
import pandas as pd
data = pd.read_csv('convertcsv.csv',delimiter=";")
listData = data.values.tolist()
print(listData)
```

```python
def dispData(data):
    print(f"NPM\t\tNAMA\t\t\t\tNILAI")
    for i in data :
        print(f"{i[0]}\t{i[1]}"+(" "*(32-len(i[1])))+i[2])
dispData(listData)
```

```python
def selectionSort(a):
    for i in range(len(a)-1):
        idx=i
        for j in range(i+1,len(a)):
            if a[idx][2] > a[j][2]:
                idx=j
        a[idx],a[i]=a[i],a[idx]
    return a
```

```python
y = listData.copy()
x = selectionSort(y)
dispData(x)
```

```python
# assending dari kiri ke kanan
def bubbleSort(data):
    for i in range(len(data)-1,0,-1):
        for j in range(i):
            if data[j][2] >data[j+1][2]:
                data[j],data[j+1]=data[j+1],data[j]
    return data
```

```python
z = listData.copy()
dispData(bubbleSort(z))
```
