### Konsep_1
#### 1. Pengertian
- Bubble Sort : algoritma bubble sort berkerja dengan cara membandingakn dua data yang berdekatan dari awal hingga akhir,dan menukar/swap data tersebut jika tidak berada pada posisi yang benar jika sudah berada pada pada posisi yang benar maka tidak di bandingkan lagi dan berganti ke selanjutnya.
- Selection Sort: algoritma selection sort hampir sama dengan bubble sort tetapi proses swaping dilakukan setelah menemukan data yang paling besar/kecil dan dilakukan swap pada posisi yang benar kemudian berganti ke selanjutnya  dan perulangan pada selection sort sama seperti bubble sort.
- Insertion Sort: algoritma insertion sort berkerja dengan cara mengasumsikan sebuah data telah terurut kemudian dibandingkan dengan flag/data yang akan di masukkan, flag ini akan di bandingkan satu persatu dengan data terurut dan di sisipkan ke data tersebut. perulangan pada insertion sort ini tidak pasti tergantung datanya.

#### 2. Program
- Bubble Sort:
~~~python
def bubbleSort(data):
    for i in range(len(data)-1,0,-1):
        print('Iterasi ke-',len(data)-i,'Jumlah Iterasi =',i)
        for j in range(i):
            if data[j] > data[j+1]:
                data[j],data[j+1]=data[j+1],data[j]
            print(data)
        print()
    return data                   
~~~

- Selection Sort:
~~~python
def selectionSort(data):
    for i in range(len(data)-1,0,-1):
        print('Iterasi ke-',len(data)-i,'Jumlah Iterasi =',i)

            if data[idxMax] < data[j]:
~~~

````python
def dispData(data):
    print("\t NPM"," "*(29-4),"NAMA"," "*(7-5),"NILAI")
    for i in data :
        spasi = (" "*(29-len(i[1])))
        spasi2 = (" "*(7-len(i[2])))
        print(i[0],spasi,i[1],spasi2,i[2])

dispData(listData)
````

````python
def insertionSort(data):
    for i in range (1,len(data)):
        idx =  i
        flag = data[idx]
        j = idx-1
        while j > -1 and flag[2] < data[j][2]:
            data[j+1] = data[j]
            j-=1
        data[j+1] = flag
        
    return data
````

Gambar 2: Insertion Sort - Ascending

````python
# Implementasi #2
def insertionSort(data):
    print(data)
    for i in range (len(data)-2,-1,-1):
        idx =  i
        flag = data[idx]
        print('falg =',flag,'ind =',idx)
        j = idx+1
        while j < len(data) and flag < data[j]:
            print("cek dengan indeks ke-",j)
            data[j-1] = data[j]
            print('inner loop',data)
            j+=1
        data[j-1] = flag
        print('data sementara',data,'\n')
        
    return data
        
a = [10,2,5,8,1,20,0,6,4]
print('sorted=',insertionSort(a))
````
