### 3. Fungsi

    ```
    def penjumlahan(a,b):

    ```

Buatlah histogram karakter (jumlah atau frekuensi masing-masing karakter, termasuk spasi) dari suatu kata. Gunakan dictionary untuk menyimpan histogram ini, contoh output dapat dilihat pada Gambar 2

(a)	kata : Mata Kuliah Struktur Data

(b)	kata : Universitas Trunojoyo Madura<br>
Gambar 2: Histogram Karakter

````python
# Jawaban 2.1.2

def histogram(strChr):
    dictHisto = {}
    print("Histogram '",strChr,"' :")
    strChr = strChr.lower()
    for i in strChr:
        if i != " ":
            dictHisto[i] = strChr.count(i)
    return dictHisto

print(histogram('Mata Kuliah Struktur Data'))
print(histogram('Universitas Trunojoyo Madura'))
````

### 2.2	Persamaan Matematika
#### 2.2.1	NPM Genap


Buatlah Fungsi dan code yang diperlukan untuk memanggil fungsi tersebut dengan menggunakan bahasa Python, antara lain :
- Fungsi createList untuk membentuk suatu list dengan parameter atau argumen berupa ukuran list 
- Fungsi untuk menghitung Persamaan (2) yaitu :
$$ y=\sum_{i=0}^{n-1} x_i $$ 
$$ P_i=\frac{1}{y}x_i $$
dimana n adalah jumlah data yang terdapat pada list (dari fungsi createList), xi adalah data ke-i dari suatu list, dan Pi adalah probabilitas dari data xi. 
<br><br>
- Fungsi untuk menghitung Persamaan Gini, seperti yang dituliskan pada Persamaan (3) yaitu:
$$ gini=1-\sum_{i=0}^{n-1}(P_i)^2 $$
dimana n adalah jumlah data yang terdapat pada list (dari fungsi createList), dan xi adalah data ke-i dari suatu list, dan Pi adalah probabiltas pada Persamaan(2) contoh output dapat dilihat pada Gambar 3



````python
#Jawaban 2.2.1

def createList(size):
    arr = []
    for i in range(size):
        inp = int(input('Masukkan data list : '))
        arr.append(inp)
    return arr

def persamaanY(arr):
    y=0
    for i in arr:
        y+=i
    return y

def probabilitas(arr,y):
    pi = [(1/y)*i for i in arr]
    return pi

def gini(pi):
    result = 0
    for i in pi :
        result+=i**2
    return 1 - result

a = createList(6)
y1 = persamaanY(a)
pi1 = probabilitas(a,y1)
gini1 = gini(pi1)

print("x2 = ",a," ; ",y1)
print("Probabilitas2 = ",pi1)
print("Gini2 = ",gini(pi1))

b = createList(1)
y2 = persamaanY(b)
pi2 = probabilitas(b,y2)
gini2 = gini(pi2)

print("x1 = ",b," ; ",y2)
print("Probabilitas1 = ",pi2)
print("Gini1 = ",gini(pi2))
````

### 2.3	Matriks1D
Buatlah Fungsi dan code untuk memanggil fungsi dengan menggunakan Python, antara lain :
- Fungsi pembuatan matriks 1D untuk membentuk suatu list 1D dengan parameter atau argumen berupa ukuran list

Gambar 5: Matriks1D
- Fungsi penjumlahan matriks 1D untuk menghitung penjumlahan dua buah matriks 1D (dengan syarat, ukuran atau jumlah data pada matriks adalah sama)
- Fungsi perkalian antara skalar dan matriks 1D untuk menghitung operasi perkalian antara sebuah bilangan dengan sebuah matriks 1D
- Fungsi perkalian dua buah matriks 1D untuk menghitung operasi perkalian antara setiap bilangan pada matriks 1D (dengan syarat, ukuran atau jumlah data pada matriks adalah sama) Contoh output yang dihasilkan dapat dilihat pada Gambar 5.


````python
#Jawaban 2.3
def createMatrix1D(size):
    mat = []
    for i in range(size):
        inp = int(input('Masukkan data mat : '))
        mat.append(inp)
    return mat

def penjumlahanMatriks1D(mat1,mat2):
    if len(mat1)==len(mat2):
        hasil=[]
        for i in range(len(mat1)):
            hasil.append( mat1[i]+mat2[i])
        return hasil
    else:
        return 'Ukuran matriks Tidak Sama'
    
def perkalianSkalar(num,mat):
    hasil = [i*num for i in mat]
    return hasil

def perkalianMatriks1D(mat1,mat2):
    if len(mat1)==len(mat2):
        hsl=[]
        for i in range(len(mat1)):
            hsl.append( mat1[i]*mat2[i])
        return hsl
    else:
        return 'Ukuran Matriks Tidak Sama'
````

````python
a = createMatrix1D(4)
print('matriks 1 = ',a)
b = createMatrix1D(3)
print('matriks 2 = ',b)
c = createMatrix1D(4)
print('matriks 3 = ',c)

jumlah1 = penjumlahanMatriks1D(a,b)
print('Hasil Penjumlahan = ',jumlah1)
print('matriks 1 = ',a)
print('matriks 3 = ',c)
jumlah2 = penjumlahanMatriks1D(a,c)
print('Hasil Penjumlahan = ',jumlah2)
perkalian1 = perkalianSkalar(3,jumlah2)
print('Hasil Perkalian skalar = ',perkalian1)
perkalian2 = perkalianMatriks1D(perkalian1,a)
print('Hasil Perkalian Matriks = ',perkalian2)
````

### 2.4	Matriks2D
Buatlah Fungsi dan Code untuk memanggil fungsi dengan menggunakan Python, antara lain :
- Fungsi pembuatan matriks 2D untuk membentuk suatu list 2D dengan parameter atau argumen berupa ukuran list (yaitu jumlah baris dan kolom), seperti yang ditunjukkan pada Gambar 6a
- Fungsi penjumlahan dua buah matriks 2D untuk menghitung penjumlahan dua buah matriks 2D (dengan syarat, ukuran matriks adalah sama, yaitu kedua matriks memiliki jumlah baris dan jumlah kolom yang sama), seperti yang ditunjukkan pada Gambar 6b
- Fungsi perkalian dua buah matriks 2D untuk menghitung operasi perkalian matriks (dengan syarat, jumlah kolom pada matriks pertama sama dengan jumlah baris pada matriks kedua), seperti yang ditunjukkan pada Gambar 6c

(a)	Pembuatan Matriks 2D

(b)	Penjumlahan Matriks 2D

(c)	Perkalian Matriks 2D
Gambar 6: Matriks 2 Dimensi


````python
#Jawaban 2.4

def createMatrix2D(baris,kolom):
    mat2D = []
    for i in range(baris):
        lst = []
        for j in range(kolom):
            inp = int(input(f'Mat[{i},{j}] = '))
            lst.append(inp)
        mat2D.append(lst)
    return mat2D

def addMatrix(mat1,mat2):
    if len(mat1) == len(mat2) and len(mat1[0]) == len(mat2[0]):
        result = []
        for i in range(len(mat1)):
            lst1 = []
            for j in range(len(mat1[0])):
                lst1.append(mat1[i][j]+mat2[i][j])
            result.append(lst1)
        return result
    else:
        return 'Ukuran Tidak Sama'

def mulMatrix2D(mat1,mat2):
    if len(mat1[0]) == len(mat2):
        hasil = []
        for i in range(len(mat1)):
            lst2=[]
            for j in range(len(mat2[0])):
                temp=0
                for k in range(len(mat1[0])):
                    temp+= mat1[i][k] * mat2[k][j]
                lst2.append(temp)
            hasil.append(lst2)
        return hasil
    else:
        return 'Ukuran Tidak memenuhi'
````

````python
mat1 = createMatrix2D(2,3)
print('mat1=',mat1)
mat2 = createMatrix2D(2,3)
print('mat2=',mat2)
mat3 = createMatrix2D(2,2)
print('mat3=',mat3)

jumlah1 = addMatrix(mat1,mat2)
jumlah2 = addMatrix(mat1,mat3)
print('mat1+mat2=',jumlah1)
print('mat1+mat3=',jumlah2)

kali1 = mulMatrix2D(mat1,mat2)
kali2 = mulMatrix2D(mat1,mat3)
kali3 = mulMatrix2D(mat3,mat2)
print('mat1*mat2=',kali1)
print('mat1*mat3=',kali2)
print('mat2*mat3=',kali3)
````
