```python
def insertionSort(data):
    for i in range(len(data)-2,-1,-1):
        print(i)
        key=data[i]
        idx=i
        print(data)
        while idx < len(data) and data[idx+1] < key:
            data[idx] = data[idx+1]
            print('inner',data)
            idx+=1
        data[idx]=key
insertionSort([11, 5, 7, 4, 5, 3, -1, 1, 13, 2, 42])
```

```python
# assending dari kanan ke kiri
def bubbleShort(listAgk):
    print('Data yang akan di urutkan',listAgk)
    for i in range(len(listAgk)-1,0,-1) :
        print('Iterasi ke-',len(listAgk)-i,'jumlah Iterasi = ',i)
        for j in range(len(listAgk)-1,len(listAgk)-i-1,-1):
#             print(i,j)
            if listAgk[j] < listAgk[j-1]:
                listAgk[j],listAgk[j-1] = listAgk[j-1],listAgk[j]
            print(j,'=>',listAgk)
    print('Data Terurut',listAgk)
    
bubbleShort([1,10,2,3,75,-7,4,5,-4])
        
```

```python
#bubblesort decending dari kanan ke kiri
def bubbleShort(listData):
    print('Data yang akan di urutkan',listData)
    count=0
    for outIter in range(len(listData)-1,-1,-1):
        print('outIter',outIter)
        count+=1
        print('iterasi ke',count)
        for i in range(len(listData)-1,len(listData)-outIter-1,-1):
            if listData[i] > listData[i-1]:
                temp =  listData[i]
                listData[i]=listData[i-1]
                listData[i-1]=temp
#                 listData[i],listData[i+1]=listData[i+1],listData[i]
            print(i,'=',listData)
    print('Data urut-',listData)
bubbleShort([-2,100,11,51,7,13,2,-1,70])
```

```python
def modifBubbleShort(listAgk):
    print('Data yang akan di urutkan',listAgk)
    i = len(listAgk)-1
    while i > 0:
        print('Iterasi ke-',len(listAgk)-i,'Jumlah Iterasi = ',i)
        temp=0
        for j in range(i):
            if listAgk[j] > listAgk[j+1]:
                listAgk[j+1],listAgk[j] = listAgk[j],listAgk[j+1]
                temp+=1
            print(j,'=>',listAgk)
        i-=1
        if temp == 0 :
            i=0
    print('Data Terurut',listAgk)
    
modifBubbleShort([1,10,2,2,2,3,4,5,2])
        
```

```python
def bidirectional(lst):
    kiri = 0
    kanan= len(lst)-1
    count=1
    while kiri < kanan:
        print('#',count)
        for i in range(kiri,kanan):
            if lst[i]>lst[i+1]:
                lst[i],lst[i+1]=lst[i+1],lst[i]
            print('kanan',lst)
        kanan-=1
        for j in range(kanan,kiri,-1):
            if lst[j] < lst[j-1]:
                lst[j],lst[j-1]=lst[j-1],lst[j]
            print('kiri',lst)
        kiri+=1
        count+=1
    return lst
print(bidirectional([100,3,7,3,6,4,90,0]))   
```
